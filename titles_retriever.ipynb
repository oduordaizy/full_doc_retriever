```python
#Importing Libraries


import numpy as np
import pandas as pd
import nltk
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.neighbors import KDTree

```

```python
#reading the csv file
books = pd.read_csv('/content/dataset1.csv')
```

```python
books
```

Creating the information retrieval system

```python
#creating the vectors
tfidf = TfidfVectorizer()
train_tfidf = tfidf.fit_transform(books.Title)
```

```python
books['tfidf'] = list(train_tfidf.toarray())
```

```python
books
```

```python
kdtree = KDTree(train_tfidf.toarray(), leaf_size = 3)
```

```python
query_tfidf = tfidf.transform(['neural'])  # Transform the query text into TF-IDF vector
distance, idx = kdtree.query(query_tfidf.toarray(), k=4)
```

```python
for i, value in list(enumerate(idx[0])):
  print(f"Book Name: {books['Title'][value]}")
  print(f"Author: {books['Author'][value]}")
```
