```python
import os
import requests
from urllib.parse import urlparse
import pandas as pd

# Create a folder if it doesn't exist
folder_name = 'documents'
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

def generate_filename(url):
    # Parse the URL to get the filename
    parsed_url = urlparse(url)
    filename = os.path.basename(parsed_url.path)
    return filename

def save_pdf(url, folder):
    # Generate a unique filename
    filename = generate_filename(url)
    file_path = os.path.join(folder, filename)

    # Check if file_path already exists and handle the case
    if os.path.exists(file_path):
        print(f"File '{file_path}' already exists. Skipping...")
        return

    # Download the PDF
    try:
        response = requests.get(url)
        response.raise_for_status()  # Raise an error for 4xx and 5xx status codes
        with open(file_path, 'wb') as f:
            f.write(response.content)
        print(f"Downloaded: {file_path}")
    except Exception as e:
        print(f"Failed to download '{url}': {e}")


# Load document URLs and filenames from the CSV file
csv_file = '/content/dataset1.csv'
document_urls = pd.read_csv(csv_file)

#Data Preprocessing
# Remove rows containing NaN values
#document_urls = document_urls.dropna()

# Write the filtered DataFrame back to the CSV file
#document_urls.to_csv('/content/filtered_dataset.csv', index=False)

# Assuming your CSV has columns 'URL' containing the PDF URLs
pdf_urls = document_urls['URL'].tolist()

# Scraping and saving each PDF document
print("The documents scraped are saved in documents folder\n")
for url in pdf_urls:
    save_pdf(url, folder_name)
```

```python
!pip install fitz
from langchain.schema import Document
from langchain.vectorstores import Chroma
from langchain.retrievers import ParentDocumentRetriever

## Text Splitting & Docloader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.storage import InMemoryStore
from langchain.document_loaders import PyPDFLoader
```

```python
!pip install pypdf
from langchain.document_loaders import PyPDFLoader
# Specify the folder path
folder_path = 'documents'

# List all PDF files in the folder
pdf_files = [file for file in os.listdir(folder_path) if file.endswith('.pdf')]

# Debug print to check the list of PDF files
print("PDF files found:", pdf_files)

# Create loaders for each PDF file
loaders = []
for pdf_file in pdf_files:
    pdf_file_path = os.path.join(folder_path, pdf_file)
    loader = PyPDFLoader(pdf_file_path)
    loaders.append(loader)

# Load documents into a list
docs = []
for loader in loaders:
    docs.extend(loader.load())

# Print the number of documents loaded
print("Number of documents loaded:", len(docs))

```

```python
from langchain.embeddings import HuggingFaceBgeEmbeddings

model_name = "BAAI/bge-small-en-v1.5"
encode_kwargs = {'normalize_embeddings': True} # set True to compute cosine similarity

bge_embeddings = HuggingFaceBgeEmbeddings(
    model_name=model_name,
    model_kwargs={'device': 'cpu'},
    encode_kwargs=encode_kwargs
)
```

```python
# This text splitter is used to create the child documents
child_splitter = RecursiveCharacterTextSplitter(chunk_size=400)


# The vectorstore to use to index the child chunks
vectorstore = Chroma(
    collection_name="full_documents",
    embedding_function=bge_embeddings  #OpenAIEmbeddings()
)

# The storage layer for the parent documents
store = InMemoryStore()

full_doc_retriever = ParentDocumentRetriever(
    vectorstore=vectorstore,
    docstore=store,
    child_splitter=child_splitter,
)
```

```python
#Checking the content in docs
# From the output, the path has been taken as a string rather than as a path
for doc in docs:
    if isinstance(doc, str):
        # Handle string differently (if needed)
        print("String document:", doc)
    else:
        # Assume it's an object with 'page_content' attribute
        print("Document with page_content:", doc.page_content)

full_doc_retriever.add_documents(docs, ids=None)
```

```python
query = input("Enter you query here: ")
```

```python
index = vectorstore.similarity_search(query, k = 9)
```

```python
retrieved_docs = full_doc_retriever.get_relevant_documents(query)

print(retrieved_docs)
```
